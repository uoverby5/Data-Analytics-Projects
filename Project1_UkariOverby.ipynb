```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv('graduation_rate.csv')
print(df)
```

This Dataset explores graduation rates over time based on certain criteria. It includes various factors such as ACT scores, SAT scores, and parental income. 

https://www.kaggle.com/datasets/rkiattisak/graduation-rate

Main inquiry question: 
1. How do these various factors influence graduation rates at colleges/universities.

SubQuestions: 
1. How does socioeconomic status correlate with years to graduate over time?
2. To what extent does ACT and SAT scores correlate to college GPA?
3. Is there a significant relationship between parental income and  college GPA?

```python
#Question2

columns = ['ACT composite score', 'SAT total score', 'parental income','college gpa', 'years to graduate']
df_columns = df[columns]
print(df_columns)
```

```python
#Question3

x = df['parental income']
y = df['years to graduate']

plt.scatter(x, y, color='blue', marker='*', alpha=0.5)
plt.xlabel('Parental Income')
plt.ylabel('Years to Graduate')
plt.title('Parental Income vs. Years to Graduate')
plt.grid(True)
plt.show()
```

Based on the scatter plot above, there is not strong correlation between parental income and years to graduate. You can see that students with different parental income amounts all take varying amount of times to graduate. Since they are so scattered, there could be other influences that contribute to the time it took to graduate. The outlier here would be the student with a parental income of $90,000+ who took 10 years to graduate.  

```python
act_scores = [25, 30, 32, 28, 26, 34, 29, 31, 33, 27]
sat_scores = [1200, 1350, 1420, 1280, 1220, 1510, 1290, 1380, 1460, 1250]
gpa = [3.0, 3.5, 3.8, 3.3, 3.1, 3.9, 3.2, 3.6, 3.7, 3.0]

plt.scatter(act_scores, gpa, label='ACT Score', color='blue', marker='*', alpha=0.5)
plt.scatter(sat_scores, gpa, label='SAT Score', color='red', marker='o', alpha=0.5)
plt.xlabel('ACT/SAT Score')
plt.ylabel('College GPA')
plt.title('Correlation between ACT/SAT Scores and College GPA')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

Based on the scatter plot above, there is a positive correlation between college gpa and ACT/SAT scores. The higher the student's GPA is the higher they scored on the ACT/SAT. There are no outliers.
